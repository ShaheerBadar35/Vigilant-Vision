```python
from tensorflow.keras.models import load_model

model = load_model("/kaggle/input/m3/tensorflow2/default/1/LSTM_with_attention_final1.h5", compile=False)
```

```python
import cv2
import numpy as np

def extract_frames(video_path, sequence_length=15):
    cap = cv2.VideoCapture(video_path)
    frames = []
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        resized_frame = cv2.resize(frame, (64, 64))
        normalized_frame = resized_frame / 255.0
        frames.append(normalized_frame)

    cap.release()

    # Divide into sequences
    sequences = []
    for i in range(0, len(frames) - sequence_length + 1, sequence_length):
        sequence = frames[i:i + sequence_length]
        sequences.append(np.array(sequence))

    return sequences
```

```python
CLASSES_LIST = ["Abuse", "Arson", "Arrest", "Explosion", "Fighting", "RoadAccidents", "Robbery", "Shooting"]

video_path = "/kaggle/input/videos1/1.mp4"  # Update with correct path

sequences = extract_frames(video_path)

for i, sequence in enumerate(sequences):
    input_tensor = np.expand_dims(sequence, axis=0)  # Shape: (1, 15, 64, 64, 3)
    predictions = model.predict(input_tensor)
    predicted_class = CLASSES_LIST[np.argmax(predictions)]
    
    print(f"Sequence {i+1}: {predicted_class} (Confidence: {np.max(predictions):.2f})")
```
